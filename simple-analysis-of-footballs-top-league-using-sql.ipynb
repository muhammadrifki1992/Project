```python
# Leagues and Their Countries

country_of_leagues = pd.read_sql(
                        """
                        SELECT Country.name AS Country_name,
                               League.name AS League_name,
                               season,
                               count(distinct stage) AS number_of_stages,
                               count(distinct HT.team_api_id) AS number_of_teams
                        FROM Match
                        JOIN Country on Country.id = match.country_id
                        JOIN League on League.id = match.league_id
                        LEFT JOIN Team AS HT on HT.team_api_id = match.home_team_api_id
                        LEFT JOIN Team AS AT on AT.team_api_id = match.away_team_api_id
                        GROUP BY Country.name, League.name, season
                        ORDER BY Country.name, League.name, season DESC
                        """,conn
                        )

country_of_leagues
```

```python
# Top leagues by Average Goals in 2015/2016 season

top_leagues_goals = pd.read_sql(
                                """
                                WITH temp_list AS
                                (
                                 SELECT league_id,
                                        League.name AS League_name,
                                        Country.name AS Country_name,
                                        season,
                                        COUNT(distinct stage) AS number_of_stages,
                                        COUNT(distinct HT.team_api_id) AS number_of_teams,
                                        COUNT(distinct stage) * 10 AS mathes_played,
                                        SUM(home_team_goal) AS total_home_team_goal,
                                        SUM(away_team_goal) AS total_away_team_goal,
                                        SUM(home_team_goal + away_team_goal) AS total_goal
                                  FROM Match
                                  JOIN Country on Country.id = match.country_id
                                  JOIN League on League.id = match.league_id
                                  LEFT JOIN Team AS HT on HT.team_api_id = match.home_team_api_id
                                  LEFT JOIN Team AS AT on AT.team_api_id = match.away_team_api_id
                                  WHERE season = "2015/2016"
                                  GROUP BY league_id
                                  HAVING number_of_stages = 38
                                 )
                                 SELECT league_id,
                                        League_name,
                                        Country_name,
                                        season,
                                        number_of_stages,
                                        number_of_teams,
                                        mathes_played,
                                        total_home_team_goal,
                                        total_away_team_goal,
                                        total_goal,
                                        CAST(total_goal AS float) / CAST(mathes_played AS float) AS average_goals
                                 FROM temp_list
                                 ORDER BY average_goals DESC
                                 """,conn
                              )

top_leagues_goals
```

```python
# Drop Team Mapping table if exist

conn.execute('DROP TABLE IF EXISTS Team_Mapping')

# Create table for Team Mapping

conn.execute(
             """
             CREATE TABLE Team_Mapping
             AS
             SELECT home_team_api_id AS id,
                    Team.team_long_name AS team_long_name,
                    Team.team_short_name AS team_short_name,
                    League.name AS league,
                    Country.name AS country
             FROM Match
             LEFT JOIN Team on Team.team_api_id = Match.home_team_api_id
             LEFT JOIN League on League.id = Match.league_id
             LEFT JOIN Country on Country.id = Match.country_id
             """
            )
```

```python
team_mapping = pd.read_sql('SELECT * FROM Team_Mapping',conn)

team_mapping
```

```python
# Top 5 Fastest Build-up Play Teams in a match of Premier League 2011/12 season

fastest_buildup_play = pd.read_sql(
                                   """
                                   SELECT DISTINCT
                                          Team.team_api_id AS id,
                                          Team.team_long_name,
                                          Team.team_short_name,
                                          strftime('%d-%m-%Y', TA.date) AS date,
                                          TM.league AS league,
                                          TA.buildUpPlaySpeed AS build_up_play_speed,
                                          TA.buildUpPlaySpeedClass AS build_up_play_speed_class
                                   FROM Team
                                   LEFT JOIN Team_Attributes AS TA on TA.team_api_id = Team.team_api_id
                                   LEFT JOIN Team_Mapping AS TM on TM.id = Team.team_api_id
                                   WHERE league = "England Premier League" AND (date BETWEEN '2011-08-13' AND '2012-05-13')
                                   ORDER BY buildUpPlaySpeed DESC
                                   LIMIT 5
                                   """,conn
                                  )

fastest_buildup_play
```

```python
# Top 5 Teams with Lowest Defences Pressure in a match of Serie A 2014/15 season

fastest_buildup_play = pd.read_sql(
                                   """
                                   SELECT DISTINCT
                                          Team.team_api_id AS id,
                                          Team.team_long_name,
                                          Team.team_short_name,
                                          strftime('%d-%m-%Y', TA.date) AS date,
                                          TM.league AS league,
                                          TA.defencePressure AS defence_pressure,
                                          TA.defencePressureClass AS defence_pressure_class
                                   FROM Team
                                   LEFT JOIN Team_Attributes AS TA on TA.team_api_id = Team.team_api_id
                                   LEFT JOIN Team_Mapping AS TM on TM.id = Team.team_api_id
                                   WHERE league = "Italy Serie A" AND (date BETWEEN '2014-08-30' AND '2015-05-31')
                                   ORDER BY defence_pressure ASC
                                   LIMIT 5
                                   """,conn
                                  )

fastest_buildup_play
```

```python
# Top 10 Players based on Overall Rating in 2015

top_players = pd.read_sql(
                          """
                          SELECT Player.player_api_id AS id,
                                 player_name,
                                 AVG(PA.overall_rating) AS overall_rating,
                                 strftime('%d-%m-%Y', Player.birthday) AS birthday,
                                 height,
                                 weight
                          FROM Player
                          LEFT JOIN Player_Attributes AS PA on PA.player_api_id = Player.player_api_id
                          WHERE PA.date BETWEEN '2015-01-01' AND '2015-12-31'
                          GROUP BY Player.player_api_id
                          ORDER BY overall_rating DESC
                          LIMIT 10
                          """,conn
                         )

top_players
```

```python
# Top 10 Potential Young Players (u-23) in 2015

top_young_players = pd.read_sql(
                                """
                                SELECT Player.player_api_id AS id,
                                       player_name,
                                       AVG(PA.potential) AS potential,
                                       AVG(PA.overall_rating) AS overall_rating,
                                       strftime('%d-%m-%Y', Player.birthday) AS birthday,
                                       height,
                                       weight
                                FROM Player
                                LEFT JOIN Player_Attributes AS PA on PA.player_api_id = Player.player_api_id
                                WHERE birthday > '1991-12-31' AND(PA.date BETWEEN '2015-01-01' AND '2015-12-31')
                                GROUP BY Player.player_api_id
                                ORDER BY potential DESC
                                LIMIT 10
                                """,conn
                               )

top_young_players
```
